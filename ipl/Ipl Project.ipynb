## Making New Column For Status of Player

```python
teams = data[data["COST IN ₹ (CR.)"]>0]["Team"].unique()
teams
```

```python
data["Status"]=data["Team"].replace(teams,"sold")
```

```python
data
```

## Players sold and unsold using bar graph

```python
plt.figure(figsize=(10,5))
fig=sns.countplot(x=data['Status'],palette=["Orange","Pink"])
plt.xlabel('Sold or Unsold')
plt.ylabel('Number of Players')
plt.title("Sold V/S Unsold")

```

```python
data.groupby("Status")["Player Name"].count()
```

```python
plt.figure(figsize=(14,7))
fig1=sns.countplot(x=data[data['Team']!="Unsold"]['Team'])
plt.xlabel("Teams Name ")
plt.ylabel("Number Of Players")
plt.xticks(rotation=70)
```

## Making new Column for Checking Retention 

```python
data["Retention"]=data["Base Price"]
```

```python
ret=data[data["Base Price"]!="Retained"]["Base Price"].unique()
```

```python
data["Retention"].replace(ret,"From Auction",inplace=True)
```

```python
data
```

```python
data.groupby(["Team","Retention"])["Retention"].count()[:-1]
```

```python
plt.figure(figsize=(12,6))
fig2=sns.countplot(x=data[data["Team"]!="Unsold"]["Team"],hue=data["TYPE"])
plt.xticks(rotation=60)
plt.title("Players In Each Team",fontsize=12)
plt.xlabel("Team Names")
plt.ylabel("Number OF Players")
```

## Total Money Spend In IPL 2023 Auction By Teams in Cr.
