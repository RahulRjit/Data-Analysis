# AQI Analysis

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns

```

```python
#importing Dataset firstly from device using pandas
data=pd.read_csv("AQI.csv")
data
```

## Now Droping Unwanted or Not Filled Data From Dataset

```python
data.dropna(axis=1,inplace=True)
```

```python
data.tail(10)
```


## Ploting Histogram For the Distribution Of AQI Values

```python
plt.figure(figsize=(14,7))
sns.distplot(data["AQI Value"])
plt.title("Distribution of AQI Values")
```

## Ploting Graph for AQI Category

```python
plt.figure(figsize=(14,7))
sns.countplot(data=data,x="AQI Category")

plt.title("AQI Category Distribution")
plt.xticks(rotation=30)
```

```python
types=data["PM2.5 AQI Category"].value_counts()
types
```

```python
plt.figure(figsize=(12, 8))
plt.pie(types.values,labels=types.index,autopct="%.2f%%",startangle=-5,)
plt.tight_layout()
```

```python
data.columns
```

## Now Will Define Relationship B/w All Values 

```python
plt.figure(figsize=(14,7))
sns.scatterplot(data=data,x="AQI Value",y="PM2.5 AQI Value")
plt.title("Relationship B/W AQI Values And PM2.5 AQI Value")
```

Relationship B/W CO AQI Value and NO2 AQI Value

```python
plt.figure(figsize=(14,7))
sns.scatterplot(x="CO AQI Value",y="NO2 AQI Value",data=data)
plt.title("Relationship B/W CO AQI Value and NO2 AQI Value")
```

## AQI Category vs. AQI Value

```python
plt.figure(figsize=(14,8))
sns.catplot(x="AQI Category",y="AQI Value",kind="box",data=data)

plt.xticks(rotation=35)
```

## Geographic distribution of cities

```python
plt.figure(figsize=(14,7))
sns.scatterplot(x="lng",y="lat",data=data,alpha=0.5)
plt.title("Geographic distribution of cities")
```

## Comparison of AQI Categories by Pollutant

```python
plt.figure(figsize=(14,8))
sns.violinplot(x="AQI Category",y="AQI Value",data=data)
plt.title("Comparison of AQI Categories by Pollutant")
plt.xticks(rotation=25)
```

## Correlation Matrix of Pollutant AQI Values

```python
#Firstly We Will Findout The Correlation In the Dataset
c=data[['AQI Value','CO AQI Value','Ozone AQI Value','NO2 AQI Value','PM2.5 AQI Value']].corr()
c
```

### Now Ploting Graph On Above Values

```python
plt.figure(figsize=(14,7))
sns.heatmap(c,xticklabels=c.columns,yticklabels=c.columns,annot=True)
```

```python
sns.pairplot(data[['AQI Value','CO AQI Value','Ozone AQI Value','NO2 AQI Value','PM2.5 AQI Value']])
```
